```python
import pandas
```

```python
datas=pandas.read_excel("图书目录.xlsx")
```

```python
datas["书名"]=datas["书名"].str.strip()
```

```python
datas["图书分类号"]=datas["图书分类号"].str.strip()
```

```python
datas[datas["书名"].notnull()]
```

```python
datas=datas[datas["书名"].notnull() &  datas["图书分类号"].notnull()]
```

```python
datas=datas[["书名","图书分类号"]]
```

```python
datas
```

```python
datas.drop_duplicates()
```

```python
datas=datas.drop_duplicates()
```

```python
datas
```

```python
datas=datas.reindex()
```

```python
datas
```

```python
#jieba.analyse.extract_tags提取书名当中的主题词
import jieba.analyse
aaa=datas.drop("书名",axis=1).join(pandas.DataFrame(jieba.analyse.extract_tags(x) for x in datas["书名"]).stack().reset_index(level=1,drop=True).rename("词条"))
```
