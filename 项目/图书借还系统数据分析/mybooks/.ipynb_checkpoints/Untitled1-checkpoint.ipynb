```python
import pandas
```

```python
books=pandas.read_excel("图书目录.xlsx")
```

```python
books
```

```python
borrow=pandas.read_excel("图书借还.xlsx")
```

```python
borrow
```

```python
author=pandas.read_excel("读者信息.xlsx")
```

```python
author
```

```python
#先把books表和借还表borrow做连接,merge按列进行连接表，concat按行连接表
new_borrow=pandas.merge(borrow,books,on="图书ID")
```

```python
new_borrow
```

```python
#把新产生的new_borrow跟author表示进行合并，merge用读者ID的维度进行连接
new_borrow_author=pandas.merge(new_borrow,author,on="读者ID")
```

```python
new_borrow_author
```

#当把读者表接到借还表和图书表联表时，数据出现了不同。少了一些记录。
得到：借还表和图书表没有出入,借还表的图书都可以在图书表中找到
第二级与作者表连接，借还表有一些作者不在当前借读用户表中，借读用户新会员没有被记录
数据分析一般结构数据特点、技术、根据技术得到的结果有一些推论。

```python
new_borrow_author1=pandas.merge(new_borrow,author,on="读者ID",how="left")
```

```python
new_borrow_author1
```

```python
#注意表连接，表联接也要注意数据不完整性
#如何做分析
```

```python
#正常来讲，图书借还管理系统，当前三个表中，借还的汇总数据
#统计一年内借书多少本,还书多少本.
new_borrow_author1.groupby("操作类型").count()
```

上面的表格输出说明什么问题:借110651本书,还110969本书.如果借书多了,书就在读者手里,图书馆进书.如果还书多了,借书少了,针对性分析为什么书没有借出去.根据技术的结果展示提出参考意见.结论就是结合业务所出来的建议.

```python
#统计某一本书的借了多少,还了多少
new_borrow_author1.groupby(["书名","操作类型"]).count().sort_values("单位",ascending=False)[:10]
```

```python
结论:<小王子>书借157次,还了150,有7本借了没有还.带来图书馆需不需再进7本书,加入进书.
```
