# Superstore的家具的时间序列预测模型

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
import pandas as pd
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

## 导入数据

```python
df=pd.read_excel(r'C:\Users\telmu\Desktop\PythonJupyterNotebook\Datasets\Superstore.xls')
df.info()
```

## 提取家具的销售数据

```python
df1=df.loc[df['Category']=='Furniture']
df1=df1.sort_values(by='Order Date')
df1=df1.groupby('Order Date')['Sales'].sum().reset_index()
df1.head()
```

```python
df1=df1.set_index('Order Date')
df1.head()
```

## 当前的日期时间数据可能很难处理，因此使用每月的平均销售额

```python
df2=df1['Sales'].resample('MS').mean()
df2.head()
```

```python
df2.plot(figsize=(10,5))
```

## 这数据有季节性规律，如年初销售总是低，年末销售总是高。
## 所以继续看它的趋势、季节性和噪声

```python
season=sm.tsa.seasonal_decompose(df2, model='additive')
season.plot()
```

## 上面的清楚地表明销售额的数据是不稳定的
## 现在找ARIMA的最优超参数

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(df2,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

## 最优的是SARIMAX(1, 1, 1)x(1, 1, 0, 12)
## 拟合ARIMA模型

```python
model=sm.tsa.statespace.SARIMAX(df2,order=(1, 1, 1),seasonal_order=(1, 1, 0, 12),enforce_stationarity=False, enforce_invertibility=False).fit()
print(model.summary().tables[1])
```

## 运行模型诊断以调查有没有任何异常行为

```python
model.plot_diagnostics(figsize=(16, 8))
```

## 不是完美，但模型残差差不多是正态分布
## 现在验证模型的正确性

```python
pred=model.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci=pred.conf_int()
ax=df2['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
predicted=pred.predicted_mean
real=df2['2017-01-01':]
mse=((predicted-real) ** 2).mean()
rmse=np.sqrt(mse)
rmse
```

## 结果是很不错的
## 现在预测未来的销售额

```python
pred_uc=model.get_forecast(steps=100)
pred_ci=pred_uc.conf_int()
ax=df2.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

## 该模型清楚地反映了家具销售的季节性。
## 随着对未来的进一步预测，自然会失去信心。这反映在该模型所产生的置信区间上，即置信区间越来越大。
